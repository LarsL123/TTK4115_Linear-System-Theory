```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np

FILE_NAME1 = 'Q10R1_integral.mat'
FILE_NAME2 = 'Q10_10_10_100_100R1_integral.mat'
FILE_NAME3 = 'Q10_10_100_10_10R1_integral.mat'

FILE_LIST = [FILE_NAME1, FILE_NAME2,FILE_NAME3]

# Set time window
t_min = 0
t_max = 15

# Set up the plots only once
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
labels = ['Time', 'Pitch', 'Pitch Rate', 'Elevation Rate', 'Reference', '?1', '?2', '?3']

# Loop over all files
for fidx, FILE_NAME in enumerate(FILE_LIST):
    data = scipy.io.loadmat(FILE_NAME)
    signal_data = data["ans"]
    time = signal_data[0, :]

    # Mask time
    mask = (time >= t_min) & (time <= t_max)
    time_cropped = time[mask]

    # Plot signals 14 on the same axes for all files
    for i in range(1, 5):   # skip time index 0
        ax_idx = i - 1
        ax = axes[ax_idx // 2, ax_idx % 2]

        signal_cropped = signal_data[i, mask]

        # Add label per file
        ax.plot(time_cropped, signal_cropped, label=f"Test {fidx+1}")

        ax.set_title(labels[i])
        ax.set_xlabel("Time [s]")
        ax.set_ylabel("Amplitude")
        ax.grid(True)

# Add legends to each subplot
for ax_row in axes:
    for ax in ax_row:
        ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Load the MATLAB file
data = scipy.io.loadmat(FILE_NAME1)

# Display the keys to see what variables are in the file
print("Variables in the .mat file:")
for key in data.keys():
    if not key.startswith('__'):
        print(f"- {key}: shape {data[key].shape if hasattr(data[key], 'shape') else type(data[key])}")

# Example: define the time range
t_min = 0  # start time in seconds
t_max = 15  # end time in seconds

# Extract the data
signal_data = data['ans']
time = signal_data[0, :]

# Create a mask for the desired time range
mask = (time >= t_min) & (time <= t_max)

# Apply the mask to time
time_cropped = time[mask]

# Set up the plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
labels = ['Time', 'Pitch', 'Pitch Rate', 'Elevation Rate', 'Reference)', '?1', '?2', '?3']

for i in range(5):
    if i == 0:
        continue  # skip time for individual plots
    
    ax_idx = i - 1
    ax = axes[ax_idx // 2, ax_idx % 2]
    
    # Apply mask to signal
    signal_cropped = signal_data[i, mask]
    
    ax.plot(time_cropped, signal_cropped)
    ax.set_title(labels[i])
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Angle [rad]' if i in [2, 4] else 'Amplitude')
    ax.grid(True)



plt.tight_layout()
plt.show()
```
